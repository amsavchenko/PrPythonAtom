```python
class HashMap:
    class Entry:
        def __init__(self, key, value):
            # TODO store key and value as attributes
            self.key = key
            self.value = value
            self.next = None  # next entry

        def get_key(self):
            # TODO return key
            return self.key

        def get_value(self):
            # TODO return value
            return self.value

        def __eq__(self, other):
            # TODO реализовать функцию сравнения
            return self.key == other.key

    def __init__(self, bucket_num=64):
        # TODO Создать массив для бакетов
        self.arr = [None for i in range  (bucket_num)]

    def get(self, key, default_value=None):
        # TODO метод get, возвращающий значение, если оно присутствует, иначе default_value
        hash_value = self._get_hash(key)
        arr_index = self._get_index(hash_value)
        element = self.arr[arr_index]
        while element is not None:
            if element.get_key() == key:
                return element.get_value()
            else:
                element = element.next
        return default_value

    def put(self, key, value):
        # TODO метод put, кладет значение по ключу, в случае, если ключ уже присутствует он его заменяет
        hash_value = self._get_hash(key)
        arr_index = self._get_index(hash_value)
        entry = self.Entry(key, value)

        if self.arr[arr_index] is None:
            self.arr[arr_index] = entry
        else:
            current = self.arr[arr_index]
            while current is not None:
                if current.get_key() == key:
                    current.value = value
                    break
                else:
                    current = current.next
            if current is None:
                entry.next = self.arr[arr_index]
                self.arr[arr_index] = entry

    def __len__(self):
        # TODO Возвращает количество Entry в массиве
        counter = 0
        for i in self.arr:
            while i is not None:
                counter += 1
                i = i.next
        return counter

    def _get_hash(self, key):
        # TODO Вернуть хеш от ключа, по которому он кладется в бакет
        return hash(key)

    def _get_index(self, hash_value):
        # TODO По значению хеша вернуть индекс элемента в массиве
        return hash_value % len(self.arr)

    def values(self):
        # TODO Задание со звездочкой!
        for i in self.arr:
            while i is not None:
                yield i.get_value()
                i = i.next



    def keys(self):
        # TODO Задание со звездочкой!
        for i in self.arr:
            while i is not None:
                yield i.get_key()
                i = i.next

    def items(self):
        # TODO Задание со звездочкой!
        # Должен возвращать генератор или итератор пар ключ и значение (tuples)
        for i in self.arr:
            while i is not None:
                k = i.get_key()
                v = i.get_value()
                yield (k, v)
                i = i.next



    def __str__(self):
        # {bucket 0: key 1, value 1; ... ; key n, value n; ...; bucket m: ... key h, value h; }
        self_copy = copy.copy(self.arr)
        string = '{'
        for i in range(len(self_copy)):
            string += str(i)
            string += ': '
            if self_copy[i] is None:
                string += 'None; '
            while self_copy[i] is not None:
                if self_copy[i].key != self_copy[i].value:
                    string += str(self_copy[i].key)
                    string += ', '
                string += str(self_copy[i].value)
                string += '; '
                self_copy[i] = self_copy[i].next
        string += '}'
        return string



```

```python
import random

def test_hashmap_04():
    entries = [(5, 7), ("entries", 56), ("value", 54.), (1000, "t"), (HashMap(10), ()), ({"s": "v"}, {"v": "s"})]
    for k, v in entries:
        entry = HashMap.Entry(k, v)
        assert entry.get_key() == k
        assert entry.get_value() == v
    print("Test 4 part 01 passed")

    for i in range(len(entries)):
        entry_one = HashMap.Entry(entries[i][0], entries[i][1])
        for _ in range(10):
            j = random.randint(0, len(entries) - 1)
            p = random.randint(0, len(entries) - 1)
            entry_two = HashMap.Entry(entries[j][0], entries[p][1])
            if j == i:
                assert entry_one == entry_two
            else:
                assert entry_one != entry_two
    print("Test 4 part 02 passed")


def test_hashmap_05():
    hashmap = HashMap(10)
    assert sum(isinstance(v, list) for k, v in hashmap.__dict__.items()) == 1
    print("Test 5 part 01 passed")

    inner_list_name = [k for k, v in hashmap.__dict__.items() if isinstance(v, list)][0]
    for i in range(10):
        assert len(HashMap(i).__dict__[inner_list_name]) == i
    print("Test 5 part 02 passed")


def test_hashmap_06():
    hashmap = HashMap(2)
    entries = [(5, 7), ("entries", 56), ("value", 54.), (1000, "t"), (HashMap(10), ())]
    for k, v in entries:
        hashmap.put(k, v)
    for k, v in entries:
        assert hashmap.get(k) == v
    print("Test 6 part 01 passed")

    for _ in range(100):
        i = random.randint(0, len(entries) - 1)
        j = random.randint(0, len(entries) - 1)
        hashmap.put(i, j)
        assert hashmap.get(i) == j

    assert hashmap.get("nexit", "default") == "default"

    print("Test 6 part 02 passed")
```

```python
test_hashmap_04()
test_hashmap_05()
test_hashmap_06()
```

```python
class HashSet(HashMap):

    def get(self, key, default_value=None):
        # TODO достаточно переопределить данный метод
        return key if super(HashSet, self).get(key) is not None else default_value

    def put(self, key):
        # TODO метод put, нужно переопределить данный метод
        super(HashSet, self).put(key, key)

    def __len__(self):
        # TODO Возвращает количество Entry в массиве
        return super(HashSet, self).__len__()

    def values(self):
        for i in self.arr:
            while i is not None:
                yield i.get_key()
                i = i.next
```

```python
import requests
from bs4 import BeautifulSoup
import json
import re
```

```python
url = 'https://www.kinopoisk.ru/top/'
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

оставил как пример

```python
el = soup.find(id='top250_place_1')
print(el)
```

```python
import pandas as pd
import numpy as np
```

```python
name_rus = [0] * 250
year = [0] * 250
name_eng = [' '] * 250
rating = [0] * 250
marks_number = [0] * 250
movie_id = [0] * 250

for i in range(1, 251):
    movie = soup.find(id='top250_place_'+str(i))
    
    name_rus[i-1] = movie.find('a', {'class': 'all'}).text
    year[i-1] = name_rus[i-1][-5:-1]
    if movie.find('span', {'class': 'text-grey'}):
        name_eng[i-1] = movie.find('span', {'class': 'text-grey'}).text
    else:
        name_eng[i-1] = name_rus[i-1][:-6]
        
    movie_id[i-1] = movie.find('div', {'class': 'item'}).get('id')[movie.find('div', {'class': 'item'}).get('id').rfind('_')+1:]
    rating[i-1] = movie.find('a', {'class': 'continue'}).text
    marks_number[i-1] = movie.find('span', {'style': 'color: #777'}).text
```

```python
top250 = pd.DataFrame(index=[i for i in range(1,251)])
top250['Название'] = name_rus
top250['Год выпуска'] = year
top250['Оригинальное название'] = name_eng
top250['ID'] = movie_id
top250['Рейтинг'] = rating
top250['Количество оценок'] = marks_number
```

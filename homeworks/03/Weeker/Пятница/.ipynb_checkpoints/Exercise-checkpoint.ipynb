# Occupation

### Introduction:


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', '|')
users.head(10)
```

### Step 4. Discover what is the mean age per occupation

```python
occupations = set(users.occupation)
res = pd.Series()
for i in occupations:
    df = users.loc[users['occupation'] == i]
    res = res.append(pd.Series(float("{0:.1f}".format(df['age'].mean())), index=[i]))

res
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
occupations = set(users.occupation)
res = pd.Series()
for i in occupations:
    df = users.loc[users['occupation'] == i]
    df_male = df.loc[df['gender'] == 'M']
    ratio = len(df_male) / len(df)
    res = res.append(pd.Series(float('{0:.2f}'.format(ratio)), index=[i]))
    
res.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
occupations = set(users.occupation)
res = pd.DataFrame()
for i in occupations:
    df = users.loc[users['occupation'] == i]
    res = res.append(pd.DataFrame({'min age': df['age'].min(), 'max age': df['age'].max()}, index=[i]))
    
res
    
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
occupations = set(users.occupation)
res = pd.DataFrame()
for i in occupations:
    df = users.loc[users['occupation'] == i]
    res = res.append(pd.DataFrame({'Male': float('{0:.1f}'.format(df.loc[df['gender'] == 'M']['age'].mean())), 
                                   'Female': float('{0:.1f}'.format(df.loc[df['gender'] == 'F']['age'].mean()))}, 
                                  index=[i]))
    
res
```

### Step 8.  For each occupation present the percentage of women and men

```python
occupations = set(users.occupation)
res = pd.DataFrame()
for i in occupations:
    df = users.loc[users['occupation'] == i]
    male_percent = len(df.loc[df['gender'] == 'M']) / len(df)
    res = res.append(pd.DataFrame({'Men (%)':  float('{0:.2f}'.format(male_percent * 100)),
                                  'Women (%)': float('{0:.2f}'.format((1 - male_percent) * 100))},
                                 index=[i]))
    
res
```
